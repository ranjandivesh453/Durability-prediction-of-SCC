```python
# Run this file once in the system for installation of required Python packages 
# for successfully running gui_xgboost_sso.py

import sys
import subprocess

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# Core packages
install("numpy")
install("pandas")

# Machine Learning packages
install("scikit-learn")
install("xgboost")
install("mealpy") # For metaheuristic optimizers (WOA, PSO, AVOA, etc.)

# GUI (Tkinter usually comes pre-installed with Python, but this ensures availability)
install("tk")

# For model serialization
install("pickle-mixin")

print("All required packages have been installed successfully.")
```
